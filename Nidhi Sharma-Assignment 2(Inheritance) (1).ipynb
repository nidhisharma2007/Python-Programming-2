```python
class Car:
    def __init__(self, model, brand):
        self.model = model
        self.brand = brand
class ElectricCar(Car):
    def __init__(self, model, brand, battery_size):
        super().__init__(model, brand)
        self.battery_size = battery_size
car1 = ElectricCar("Honda", "Tesla", "85 kW")
print(car1.model, car1.brand, car1.battery_size)
```

```python
class Computer:
    def __init__(self, brand, processor):
        self.brand = brand
        self.processor = processor
class Laptop(Computer):
    def __init__(self, brand, processor, battery_life):
        super().__init__(brand, processor)
        self.battery_life = battery_life
s1 = Laptop("HP", "i7", "10 hours")
print(s1.brand, s1.processor, s1.battery_life)
```

```python
class ElectronicDevice:
    def __init__(self, brand):
        self.brand = brand
class Smartphone(ElectronicDevice):
    def __init__(self, brand, storage, operating_system):
        super().__init__(brand)
        self.storage = storage
        self.operating_system = operating_system
class Tablet(ElectronicDevice):
    def __init__(self, brand, screen_size, operating_system):
        super().__init__(brand)
        self.screen_size = screen_size
        self.operating_system = operating_system
x = Smartphone("iPhone 15 Pro", "256GB", "iOS 17")
print(x.brand, x.storage, x.operating_system)
```

```python
class Artist:
    def __init__(self, name, age):
        self.name = name
        self.age = age
class Musician(Artist):
    def __init__(self, name, age, instrument_played):
        super().__init__(name, age)
        self.instrument_played = instrument_played
x = Musician("Taylor Swift", 34, "Guitar")
print(x.name, x.age, x.instrument_played)
x2 = Musician("Ed Sheeran", 33, "Acoustic Guitar")
print(x2.name, x2.age, x2.instrument_played)
```

```python
class Book:
    def __init__(self, writer, name):
        self.writer = writer
        self.name = name
class FictionBook(Book):
    def __init__(self, writer, name, genre):
        super().__init__(writer, name)
        self.genre = genre
s1 = FictionBook("J.K. Rowling", "Harry Potter", "Fantasy")
print(s1.writer, s1.name, s1.genre)
```

```python
class student:
    def __init__(self,name,marks1,marks2,marks3):
        self.name=name
        self.marks=[marks1,marks2,marks3]
        
a1 = student("Rohan", 75, 88, 92)
total = sum(a1.marks)
length = len(a1.marks)
average = total / length
print(f"Total marks => {total} and its average is {average}")
```
