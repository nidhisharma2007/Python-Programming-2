```python
class RacingCar:
    def __init__(self):
        self.__speed = 0  

    def set_speed(self, speed):
        self.__speed = speed

    def acceleration(self, total_time):
        if total_time > 0:
            acceleration = self.__speed / total_time
            print("Acceleration:", acceleration, "m/s²")
        else:
            print("Time must be greater than 0")

    def braking(self, brake_value):
        self.__speed -= brake_value
        if self.__speed < 0:
            self.__speed = 0
        print("Speed after braking:", self.__speed)


car = RacingCar()
car.set_speed(100)
car.acceleration(10)
car.braking(30)
```

```python
class Diet:
    def __init__(self):
        self.__calories = 0

    def junk_calories(self, junk):
        self.__calories = self.__calories + junk

    def exercise(self, burn):
        self.__calories = self.__calories - burn
        if self.__calories < 0:
            self.__calories = 0

    def check_health(self):
        if self.__calories > 30:
            print("Unhealthy")
        else:
            print("Healthy")

d = Diet()
d.junk_calories(40)
d.check_health()
d.exercise(20)
d.check_health()
```

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def get_balance(self):
        return self.__balance

    def deposit(self, amount):
     if amount > 0:
        self.__balance += amount
        print("Amount deposited")
     else:
        print("Invalid deposit")

    def withdraw(self, amount):
     if amount <= self.__balance:
        self.__balance -= amount
        print("Withdrawal successful")
     else:
        print("Insufficient balance")

acc = BankAccount(1000)
acc.deposit(500)       
acc.withdraw(200)       
print(acc.get_balance())


```

```python
class IMBD:
    def __init__(self, rating):
        self.__rating = rating

    def update_rating(self, new_rating):
        if new_rating > self.__rating:
            print("Rating increased")
        else:
            print("Rating decreased")

        self.__rating = new_rating   

    def get_rating(self):
        return self.__rating


movie = IMBD(7)

movie.update_rating(9)
print(movie.get_rating())

movie.update_rating(5)   
print(movie.get_rating())

```

```python
class Laptop:
    def __init__(self, price):
        self.__price = price

  
    def set_price(self, new_price):
        if new_price > 0:
            self.__price = new_price
        else:
            print("Price must be positive.")


    def get_price(self):
        return self.__price


laptop1 = Laptop(1000)

print("Original Price:", laptop1.get_price())

laptop1.set_price(1200)
print("Updated Price:", laptop1.get_price())


```

```python
class Book:
    def __init__(self, title, author):
        self.__title = title
        self.__author = author

    def get_title(self):
        return self.__title

    def get_author(self):
        return self.__author

book1 = Book("1984", "George Orwell")

print("Title:", book1.get_title())
print("Author:", book1.get_author())

```
