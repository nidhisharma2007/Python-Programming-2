# ITERATOR

```python
fruits = ("apple", "banana", "mango", "grapes")

fruit_iterator = iter(fruits)

print(next(fruit_iterator))
print(next(fruit_iterator))
```

```python
word = "DATA"

word_iter = iter(word)

while True:
    try:
        print(next(word_iter))
    except StopIteration:
        break
```

```python
values = [12, 24, 36, 48]

value_iter = iter(values)

for v in value_iter:
    print(v)
```

```python
num_iterator = iter(range(3, 9))

print(next(num_iterator))
print(next(num_iterator))
print(next(num_iterator))
```

```python
numbers = iter([1000, 1500, 2000, 2500, 3000, 3500, 4000])

while True:
    try:
        num = next(numbers)
        if num % 2 == 0:
            print(num)
    except StopIteration:
        break
```

# GENERATOR

```python
def country_generator():
    yield "India"
    yield "Japan"
    yield "Canada"

g = country_generator()

print(next(g))
print(next(g))
print(next(g))
```

```python
def simple_generator(n):
    i = 1
    while i <= n:
        yield i
        i += 1

g = simple_generator(5)

for num in g:
    print(num)
```

```python
square_gen = (x**2 for x in range(1,6))

for s in square_gen:
    print(s)
```

```python
def odd_numbers(n):
    for i in range(1, n+1, 2):
        yield i

g = odd_numbers(10)

for num in g:
    print(num)
```

```python
def reverse_string(text):
    for i in text[::-1]:
        yield i

g = reverse_string("Python")

for i in g:
    print(i)
```

# DECORATOR

```python
def greet_decorator(func):
    def wrapper():
        print("Hello!")
        func()
    return wrapper

@greet_decorator
def say_name():
    print("My name is Alex")

say_name()
```

```python
def display_message(func):
    def wrapper():
        print("Start")
        func()
        print("End")
    return wrapper

@display_message
def show():
    print("Function is running")

show()
```

```python
def log_decorator(func):
    def wrapper(a, b):
        print("Executing function")
        return func(a, b)
    return wrapper

@log_decorator
def add(a, b):
    return a + b

print(add(3, 4))
```

```python
def check_even(func):
    def wrapper(n):
        print("Checking number")
        func(n)
    return wrapper

@check_even
def is_even(n):
    if n % 2 == 0:
        print("Number is even")
    else:
        print("Number is odd")

is_even(6)
```

```python
def decor(func):
  def wrapper(*args, **kwargs):
    result = func(*args, **kwargs)
    return result*2
  return wrapper

@decor
def get_number():
  return 5

print(get_number())
```
