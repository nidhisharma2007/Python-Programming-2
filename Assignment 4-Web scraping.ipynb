```python
import requests
from bs4 import BeautifulSoup

url = "https://www.scrapethissite.com/pages/simple/"
response = requests.get(url)

if response.status_code == 200:
    print('Success')
else:
    print('Failed to retrieve the page')


soup = BeautifulSoup(response.text, 'html.parser')

countries = soup.find_all("h3")
for country in countries:
    print(country.text)

print('Title:', soup.title.text)
print(soup.prettify())


```

```python
import requests
from bs4 import BeautifulSoup

def scrape_data(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    items = soup.find_all("a")

    for i in items:
        name = i.get_text(strip=True)
        print(name)

scrape_data("http://crawler-test.com/")
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://quotes.toscrape.com/"

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, "html.parser")

quotes = soup.find_all("span", class_="text")

for q in quotes:
    print(q.get_text(strip=True))
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.scrapingbee.com/blog/scraper-sites/"

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, "html.parser")

elements = soup.find_all(["h1", "h2", "p"])

for i, e in enumerate(elements, start=1):
    print(i, e.get_text(strip=True))
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.w3schools.com/html/html_tables.asp"
response = requests.get(url)

soup = BeautifulSoup(response.content, "html.parser")

table = soup.find("table")

rows = table.find_all("tr")

data = []

for row in rows:
    cols = row.find_all(["th", "td"])
    cols = [col.get_text().strip() for col in cols]
    data.append(cols)

for row in data:
    print(row)

print("First Row:", data[0])
print("Total Rows:", len(data))
```
